```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns 
import matplotlib.pyplot as plt 
df = pd.read_csv("./data/Data-Collisions.csv") 
```

Get the ratio of column with above 80% of missing values

```python
missing_df = df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['columns_name','missing_count']
missing_df['missing_ratio'] = missing_df['missing_count'] /df.shape[0]
missing_df.loc[missing_df['missing_ratio']>0.777]
```

```python
df.dtypes
```

```python
x_cols = [col for col in df.columns if col not in ['SEVERITYCODE', 'SEVERITYCODE.1'] if (df[col].dtype=='float64' or df[col].dtype=='int64')]

labels = []
values = []
for col in x_cols:
    labels.append(col)
    values.append(np.corrcoef(df[col].values, df.SEVERITYCODE.values)[0,1])
corr_df = pd.DataFrame({'col_labels':labels, 'corr_values':values})
corr_df = corr_df.sort_values(by='corr_values')

ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(12,40))
rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='y')
ax.set_yticks(ind)
ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal')
ax.set_xlabel("Correlation coefficient")
ax.set_title("Correlation coefficient of the variables")
plt.show()
```

we can detect that SDOTCOLUMN, INTKEY, Y,X , there is no correlation with the SEVERITYCODE

Let's get our highly correlated data

```python
corr_df_sel = corr_df.loc[(corr_df['corr_values']>0.05) | (corr_df['corr_values'] < -0.05)]
corr_df_sel
```

```python
corr_df_ = corr_df_sel.col_labels.tolist()

tem_df = df[corr_df_]

corrmat = tem_df.corr(method='spearman')
fig,ax= plt.subplots(figsize=(8,8))

sns.heatmap(corrmat,vmax=1,square = True)
plt.title('corr map',fontsize=15)
plt.show()
```

##### Decision Tree


Most of features are tyoe of object, we have to convert them to be numerical so wa could use them in our analysis

```python
df["ROADCOND"] = df["ROADCOND"].astype('category')
df["ROADCOND_cat"] = df["ROADCOND"].cat.codes

df["LIGHTCOND"] = df["LIGHTCOND"].astype('category')
df["LIGHTCOND_cat"] = df["LIGHTCOND"].cat.codes

df["ADDRTYPE"] = df["ADDRTYPE"].astype('category')
df["ADDRTYPE_cat"] = df["ADDRTYPE"].cat.codes

df["WEATHER"] = df["WEATHER"].astype('category')
df["WEATHER_cat"] = df["WEATHER"].cat.codes

df.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X = np.asarray(df[['WEATHER_cat', 'ROADCOND_cat', 'LIGHTCOND_cat', 'ADDRTYPE_cat']])
Y = np.asarray(df['SEVERITYCODE'])

from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth=7)
clf = clf.fit(X,Y)
```
